```python
import pandas as pd

# Data Preparation

countries = ["UK", "USA", "China", "India"]
carbon_intensity = [0.1, 0.2, 0.4, 0.2]
nd_gain = [70.3, 67.7, 60.2, 47.1]
baseline_e0 = [0.31, 5.08, 11.35, 2.83]

alpha_derived = [1000 * ci for ci in carbon_intensity]
d_derived = [100 / g for g in nd_gain]


# Create DataFrame with Proper Column Names
df = pd.DataFrame({
    "Country": countries,
    "CO₂ Intensity": carbon_intensity,
    "a_i (1000*CI)": alpha_derived,
    "ND-GAIN": nd_gain,
    "d_i (100/ND)": d_derived,
    "Baseline e₀ (GtCO₂)": baseline_e0
})


# Apply Styling with Different Color Gradients
styled_df = (df.style
             .format("{:.2f}", subset=["CO₂ Intensity", "a_i (1000*CI)", "ND-GAIN", "d_i (100/ND)", "Baseline e₀ (GtCO₂)"])
             .hide(axis="index")  # Hide row indices
             .background_gradient(cmap="Blues", subset=["CO₂ Intensity"])
             .background_gradient(cmap="Reds", subset=["a_i (1000*CI)"])
             .background_gradient(cmap="Greys", subset=["ND-GAIN"])
             .background_gradient(cmap="Purples", subset=["d_i (100/ND)"])
             .background_gradient(cmap="Greens", subset=["Baseline e₀ (GtCO₂)"])
             .set_properties(**{
                 'text-align': 'center',
                 'font-family': 'Arial',
                 'border': '1px solid lightgrey'
             })
             .set_caption("Figure 1: Climate Parameters by Country")  # Title: Figure 1: Climate Parameters by Country
             .set_table_styles([{
                 'selector': 'caption',
                 'props': [('font-size', '14pt'),
                           ('font-weight', 'bold'),
                           ('color', '#2e6e8e')]
             }])
            )

# Display the Dataframe
display(styled_df)
```
